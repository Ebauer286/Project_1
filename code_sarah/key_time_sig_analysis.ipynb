```python
# importing all needed libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the master data csv into a dataframe
all_songs_df = pd.read_csv(r'resources/master_data_final.csv').dropna()
all_songs_df.head()
```

```python
# rewritting key data from integers into actual key signature values
all_songs_df['key'] = all_songs_df['key'].replace(0, 'C')
all_songs_df['key'] = all_songs_df['key'].replace(1, 'C♯/D♭')
all_songs_df['key'] = all_songs_df['key'].replace(2, 'D')
all_songs_df['key'] = all_songs_df['key'].replace(3, 'D#/E♭')
all_songs_df['key'] = all_songs_df['key'].replace(4, 'E')
all_songs_df['key'] = all_songs_df['key'].replace(5, 'F')
all_songs_df['key'] = all_songs_df['key'].replace(6, 'F#/G♭')
all_songs_df['key'] = all_songs_df['key'].replace(7, 'G')
all_songs_df['key'] = all_songs_df['key'].replace(8, 'G#/A♭')
all_songs_df['key'] = all_songs_df['key'].replace(9, 'A')
all_songs_df['key'] = all_songs_df['key'].replace(10, 'A#/B♭')
all_songs_df['key'] = all_songs_df['key'].replace(11, 'B')
all_songs_df['key'] = all_songs_df['key'].replace(-1, 'no key detected')
all_songs_df.head()
```

```python
# rewritting time signature data from integers into actual time signature values
all_songs_df['time_signature'] = all_songs_df['time_signature'].replace(3, '3/4')
all_songs_df['time_signature'] = all_songs_df['time_signature'].replace(4, '4/4')
all_songs_df['time_signature'] = all_songs_df['time_signature'].replace(5, '5/4')
all_songs_df['time_signature'] = all_songs_df['time_signature'].replace(6, '6/4')
all_songs_df['time_signature'] = all_songs_df['time_signature'].replace(7, '7/4')

updated_df = all_songs_df
updated_df.head()
```

```python
# defining function to chart key signature data into a bar chart from a certain time frame
def key_filter(start, end):
    global updated_df
    keys = ['C','C♯/D♭','D','D#/E♭','E','F','F#/G♭','G','G#/A♭','A','A#/B♭','B']

    df_decade = updated_df.query(f'{start} <= year < {end}')

    count_0 = len(df_decade[(df_decade['key' ] == 'C')]['key'].to_list())
    count_1 = len(df_decade[(df_decade['key' ] == 'C♯/D♭')]['key'].to_list())
    count_2 = len(df_decade[(df_decade['key' ] == 'D')]['key'].to_list())
    count_3 = len(df_decade[(df_decade['key' ] == 'D#/E♭')]['key'].to_list())
    count_4 = len(df_decade[(df_decade['key' ] == 'E')]['key'].to_list())
    count_5 = len(df_decade[(df_decade['key' ] == 'F')]['key'].to_list())
    count_6 = len(df_decade[(df_decade['key' ] == 'F#/G♭')]['key'].to_list())
    count_7 = len(df_decade[(df_decade['key' ] == 'G')]['key'].to_list())
    count_8 = len(df_decade[(df_decade['key' ] == 'G#/A♭')]['key'].to_list())
    count_9 = len(df_decade[(df_decade['key' ] == 'A')]['key'].to_list())
    count_10 = len(df_decade[(df_decade['key' ] == 'A#/B♭')]['key'].to_list())
    count_11 = len(df_decade[(df_decade['key' ] == 'B')]['key'].to_list())

    counts = [count_0, count_1, count_2, count_3, count_4, count_5, count_6, count_7, count_8, count_9, count_10, count_11]

    plt.bar(keys, counts)
    plt.title(f"Key Distribution from {start} to {end}")
    plt.xlabel("key")
    plt.ylabel("# of songs")
    plt.savefig(f"resources/image_saves/keys_{start}-{end}.png")
```

```python
# defining function to chart time signature data into a bar chart from a certain time frame
def time_sig_filter(start, end):
    global updated_df
    time_sigs = ['3/4','4/4','5/4','6/4','7/4']

    df_decade = updated_df.query(f'{start} <= year < {end}')

    count_0 = len(df_decade[(df_decade['time_signature' ] == '3/4')]['time_signature'].to_list())
    count_1 = len(df_decade[(df_decade['time_signature' ] == '4/4')]['time_signature'].to_list())
    count_2 = len(df_decade[(df_decade['time_signature' ] == '5/4')]['time_signature'].to_list())
    count_3 = len(df_decade[(df_decade['time_signature' ] == '6/4')]['time_signature'].to_list())
    count_4 = len(df_decade[(df_decade['time_signature' ] == '7/4')]['time_signature'].to_list())

    counts = [count_0, count_1, count_2, count_3, count_4]

    plt.bar(time_sigs, counts)
    plt.title(f"Time Signature Distribution from {start} to {end}")
    plt.xlabel("time signature")
    plt.ylabel("# of songs")
    plt.savefig(rf"resources/image_saves/time_sig_{start}-{end}.png")
```

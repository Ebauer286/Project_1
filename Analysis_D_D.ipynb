```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
songs_data = pd.read_csv("master_data_final.csv")
songs_data.head()                 
```

```python
songs_average_data = pd.DataFrame(songs_data.groupby('year')['duration'].mean().round(0).astype(int)).reset_index()
```

```python
# Assuming 'x' is the column for the x-axis and 'y' for the y-axis
plt.figure(figsize=(10, 5))
plt.plot(songs_average_data['year'], songs_average_data['duration'], marker='o')
plt.title('Average Song Duration (seconds) by Year')
plt.xlabel('Year')
plt.ylabel('Duration (s)')
plt.grid()
plt.show()
```

Duration Analysis: From around 1980s to the early 1990s, there is a general increase in the average song duration, peaking at around 260 seconds (or approximately 4 minutes and 20 seconds), After the peak in the 1990s the song durations started to gradually decline when it reacheses the lowest duration, showing average durations close to 200 seconds (or about 3 minutes and 20 seconds), that compared to the highest peaks has a difference of 60 seconds.
this trend could be due to a variety of factors like streaming influence where Shorter songs are more common now, potentially influenced by streaming platforms (spotify), where shorter, catchy songs tend to perform better in terms of engagement and replay.

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='year', y='duration', data=songs_average_data)
plt.title('Songs average duration')
plt.grid()
plt.show()
```

```python
songs_dan_ave_data = pd.DataFrame(songs_data.groupby('year')['danceability'].mean()).reset_index()
```

```python
# Assuming 'x' is the column for the x-axis and 'y' for the y-axis
plt.figure(figsize=(10, 5))
plt.plot(songs_dan_ave_data['year'], songs_dan_ave_data['danceability'], marker='o')
plt.title('Average Song Danceability by Year')
plt.xlabel('Year')
plt.ylabel('Danceability')
plt.grid()
plt.show()
```

Danceability Analisys: Spotify describes danceability as how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.
It can be notice some trends over time, like the danceability seems relatively stable from 1980 to 2000 with only slight variations but there appears to be an upward trend in danceability starting around the late 2000s, with some fluctuations, having the lowest levels in 2011 and the highest around 2019 and 2020.
When we compare the data on danceability and song length, we can see that around 2020, song length had its lowest levels with shorter songs, and danceability for those same years had its highest levels. This could have been influenced by multiple factors such as social media platforms like tik-tok tahat had become a dominant force in the music industry. TikTok trends favor short, catchy, and highly danceable music snippets that can be easily incorporated into viral dance challenges or other trends. As a result, artists and producers might have focused on creating shorter songs with high danceability to maximize their impact on TikTok and similar platforms and Pandemic and Changes in Listening Habits: With the global pandemic, people were spending more time on digital platforms and streaming services, consuming music differently than they would in typical social settings. Short, upbeat, and danceable songs provided a quick mood boost and became popular in home-based activities like exercise routines, personal dance breaks, and online interactions. 
in conclusion the fact that song duration reached a low while danceability hit a high around the 2020 suggests that the music industry adapted to social media trends, changing consumption habits, and the digital, home-based lifestyle of those years.


```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='year', y='danceability', data=songs_dan_ave_data)
plt.title('Songs average danceability')
plt.grid()
plt.show()
```

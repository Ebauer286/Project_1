```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
songs_data = pd.read_csv("master_data_final.csv")
songs_data.head()                 
```

```python
songs_average_data = pd.DataFrame(songs_data.groupby('year')['duration'].mean().round(0).astype(int)).reset_index()
print(songs_average_data)
```

```python
# Assuming 'x' is the column for the x-axis and 'y' for the y-axis
plt.figure(figsize=(10, 5))
plt.plot(songs_average_data['year'], songs_average_data['duration'], marker='o')
plt.title('Average Song Duration (seconds) by Year')
plt.xlabel('Year')
plt.ylabel('Duration (s)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='year', y='duration', data=songs_average_data)
plt.title('Songs average duration')
plt.grid()
plt.show()
```

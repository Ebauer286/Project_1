```python
# importing all needed libraries
import requests
import json
import re
import pandas as pd

from spotify_token_request import r

r
```

```python
#spotify information and headers for future api use
spotify_api_key = r['access_token']
spotify_headers = {
    'Authorization': 'Bearer {}'.format(spotify_api_key)
}
```

```python
songdf = pd.read_csv('all_top100songs_by_year.csv')
song_dictionary = dict(zip(songdf['title'], songdf['artist']))
```

```python
song_dictionary
```

```python
# creating a new dictionary with edited artist values to not confuse the spotify search api, only takes the first artist when collaborations are formatted with "featured", "with", "x", etc...
search_dictionary = {}
split_artist = []

for s in song_dictionary:
    split_artist = re.split(' Featuring | With | Duet | X | x | & |, | \(| And | Feat.', song_dictionary[s])
    split_song = re.split(' \(', s)
    first_part_song = split_song[0]
    song_final = first_part_song.replace("/"," ")
    search_dictionary[song_final] = split_artist[0]
```

```python
song_id_dictionary = {}

for song in search_dictionary:
    artistsplit = search_dictionary[song].split()
    artist = '+'.join(artistsplit)
    songsplit = song.split()
    songtitle = '+'.join(songsplit)

    try:
        search_url = f"https://api.spotify.com/v1/search?q=track%3A{songtitle}+artist%3A{artist}&type=track"

        search_response = requests.get(search_url, headers = spotify_headers).json()
        song_id_dictionary[song] = search_response["tracks"]["items"][0]["id"]
#        print(f"ID for song named {song} added to list")

    except:
        song_id_dictionary[song] = ""
        print(f"no ID available for song named: {song} by {search_dictionary[song]}")
    pass
```

```python
song_id_dictionary
```

```python
id_list = []
for id in song_id_dictionary:
    id_list.append(song_id_dictionary[id])


temp_ids = dict(zip(songdf['title'], id_list))
temp_df = pd.DataFrame(temp_ids, index = [0])
temp_df = temp_df.transpose()
temp_df = temp_df.reset_index()
temp_df.columns = ["title","id"]

temp_df.head()

id_df = pd.merge(songdf, temp_df, on = "title")
id_df.head()

id_df.to_csv('all_top100songs_by_year_', index=False)
```
